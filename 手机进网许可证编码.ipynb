```python
# 写入 CSV 文件
with open('单独列的数据/手机进网许可证编号.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    for license in licenses:
        writer.writerow([license.strip()])  # 写入每个许可证编号为一行

print("数据已成功写入 licenses.csv")
```

# 测试通过，开始循环写入

```python
import os

# 设置代理环境（如果需要）
os.environ["http_proxy"] = "127.0.0.1:7890"
os.environ["https_proxy"] = "127.0.0.1:7890"

# config.py
from dotenv import load_dotenv

# 加载 .env 文件
load_dotenv()

# 读取
ZetaTechs_api_key = os.getenv('ZetaTechs_api_key')
ZetaTechs_api_base = os.getenv('ZetaTechs_api_base')

from openai import OpenAI

client = OpenAI(api_key=ZetaTechs_api_key, base_url=ZetaTechs_api_base)

system_prompt = "你是数据生成的专家，你需要根据我给出的数据样式生成一系列仿真数据。"
user_prompt = "请生成一系列类似02-8489-160510和02-0674-163860的随机的手机进网许可证编号。要求只返回你生成的进网许可证编码即可，不需要其他内容。进网许可证编号之间用逗号分隔。"

total_len = 0
required_total_len = 1025 # 生成的数据总长度

import csv

while total_len < required_total_len:

    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ]
    )

    # 假设您已经获取了生成的进网许可证编号
    data = completion.choices[0].message.content

    # 将编号按逗号分隔
    licenses = data.split(', ')

    if type(licenses) == list: # 简单增加一下容错

        # 写入 CSV 文件
        with open('单独列的数据/手机进网许可证编号.csv', mode='a', newline='') as file:
            writer = csv.writer(file)
            for license in licenses:
                writer.writerow([license.strip()])  # 写入每个许可证编号为一行

        total_len += len(licenses)

        print(f"数据已成功写入 licenses.csv。当前总行数：{total_len}")
```
